# Task 3: Analyze the distribution of gun violence incidents over the years using a line plot

```python
sd_over_years = sd.groupby("year").sum() ## To make the group by year and then calculate the sum of each group
```

```python
sd_over_years
```

```python
plt.figure(figsize=(15, 6), dpi=70, facecolor="#264653") ## Figure size
plt.xlabel("Year") ## Give uus x axis
plt.xticks(sd_over_years.index, rotation= 90) # Give us years in x axis
plt.ylabel("fatalities") ## Give us y axis
plt.title("US gun violence fatalities between 1982 - 2023") ## As we can see the increase in between 2015 to 2019
plt.plot(sd_over_years.index, sd_over_years["fatalities"]) ## For plot chart
plt.scatter(sd_over_years.index, sd_over_years["fatalities"]) ## For scattered chart in line chart

plt.show()
```

# Task 5: Calculate the total number of facilities and caused by mass shooting

```python
grouped_data = sd.groupby("type").sum()
mass_fatalities = grouped_data.loc["Mass", "fatalities"]
print(f"Total fatalities between 1982 to 2023 caused by mass shooting is: {mass_fatalities}")
```

# Task 6: Visualize the top 10 states with the highest number of incident using a bar chart

# Task 7: Determine the most common locations where mass shooting occurs

```python
location_counts = sd["location"].value_counts()
location_with_highest_incident = location_counts.idxmax()
highest_incident_count = location_counts.max()

print(f"Location with the highest incident count is {location_with_highest_incident} and {highest_incident_count} incidents.")
```

# Task 8: Calculate the average number of fatalities and injuries per incident

```python
sd["injured"] = pd.to_numeric(sd["injured"], errors="coerce") ## to chane in numeric data type
```

```python
avg_facilities = sd["fatalities"].mean()
avg_injuries = sd["injured"].mean()
print(f"The average fatalities due to gun violence between 1982-2023 is {avg_facilities}")
print(f"The average injuries due to gun violence between 1982-2023 is {avg_injuries}")
```

# Task 9: Explore the relationship between the number of fatalities and the number of injured people using a scatter plot 

```python
fatalities = sd["fatalities"]
injuries = sd["injured"]
plt.scatter(fatalities, injuries)
plt.show()
```

# Task 10: Analyze the trend of incidents based on the shooter's gender using a pie chart

```python
sd.columns
```

```python
fatalities = sd.groupby("gender").sum()["fatalities"]
labels = ["F", "F (identified as transgender)", "Female", "male", "Male & Female"]
plt.pie(fatalities)
plt.title
plt.axis("equal")
plt.legend(labels, loc="upper left")
plt.savefig("Gun Violence in U.S.A trend based on shooter's gender" format="PNG")
plt.show
```
